# 问题三：确定最小螺距，使得龙头前把手能够沿着相应的螺线盘入到调头空间的边界。
本题可以基于模型2进行些许调整。

只关注位置信息，不需要时间变量。

只需要预估大致螺距范围，内圈不碰撞，那么外圈也不会碰撞，所以从掉头空间外围一圈开始盘入即可。

```python
from model_3 import Model_3
```

```python
# 初始化模型参数
num_sections = 223  # 板凳龙节数
head_length = 2.86  # 头部长度 (m)
body_length = 1.65  # 身体每节长度 (m)  
distance = 0.55  # 初始螺距 (m)
extension = 0.275  # 连接点到窄边的距离 (m)
width = 0.3  # 宽度 (m)
r=4.5   # 掉头空间 (m)

# 创建 Model_2 实例
model = Model_3(num_sections, head_length, body_length, distance, extension, width)
```

```python
# 初筛，确定大致范围
model.min_distance(r, 0.3, 0.55, 10, 0.1)
```

```python
# 范围大致确定在0.40到0.45
model.min_distance(r, 0.4, 0.45, 5, 0.02)
```

```python
# 范围大致确定在0.4234375到0.425
model.min_distance(r, 0.4234375, 0.425, 5, 0.02)
```

因此最小螺距约为0.4243m
