# 问题五：龙头最大速度
思路：找到龙头以1m/s运动的过程中，连接点速度的最大值v0
v_h_max=2/v0

从问题4的结果分析，连接点速度最大值在14~16s之间出现，约为1.362277247m/s，下面进一步精确计算。

```python
from model_5 import Model_5

# 初始化模型参数
num_sections = 223  # 板凳龙节数
head_length = 2.86  # 头部长度 (m)
body_length = 1.65  # 身体每节长度 (m)  
distance = 1.7  # 螺距 (m)
extension = 0.275  # 连接点到窄边的距离 (m)
width = 0.3  # 宽度 (m)
r = 4.5  # 掉头区域半径 (m)
k = 2.0  # r_A/r_B

model = Model_5(num_sections, head_length, body_length, distance, extension, width, r, k)
```

```python
# 精度默认为0.01
model.trace_period(14.0, 16.0, 1)
```

在0.01精度下，运动过程中连结点最大速度为1.503365819，为使得舞龙队各把手的速度均不超过 2 m/s，龙头最大行进速度为2 ÷ 1.503365819 = 1.330348m/s
